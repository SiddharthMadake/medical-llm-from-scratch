```python
# Cell 1 — Always run this first
import torch
import torch.nn as nn
import math

class MedGPTConfig:
    vocab_size   = 16000
    block_size   = 256
    n_embd       = 256
    n_heads      = 8
    n_layers     = 6
    dropout      = 0.1
    pad_token_id = 0

cfg = MedGPTConfig()
print("✅ Imports + Config ready")
```

```python
class MedEmbedding(nn.Module):
    """
    Token Embedding  : maps token ID → dense vector
    Position Embedding: tells model WHERE each token is
    """
    def __init__(self, cfg):
        super().__init__()
        self.token_emb = nn.Embedding(cfg.vocab_size, cfg.n_embd,
                                       padding_idx=cfg.pad_token_id)
        self.pos_emb   = nn.Embedding(cfg.block_size, cfg.n_embd)
        self.drop      = nn.Dropout(cfg.dropout)
        self.block_size = cfg.block_size

    def forward(self, x):
        B, T = x.shape
        assert T <= self.block_size, f"Sequence {T} > block_size {self.block_size}"

        positions = torch.arange(T, device=x.device).unsqueeze(0)  # (1, T)
        tok = self.token_emb(x)       # (B, T, n_embd)
        pos = self.pos_emb(positions) # (1, T, n_embd)
        return self.drop(tok + pos)   # (B, T, n_embd)

# Test
emb_layer = MedEmbedding(cfg)
dummy = torch.randint(0, cfg.vocab_size, (2, 16))
out   = emb_layer(dummy)
print(f"✅ Embedding output: {out.shape}  →  (batch, seq_len, n_embd)")
```
